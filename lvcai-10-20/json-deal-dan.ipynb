```python
import json
import csv
import os
import numpy as np
import pandas as pd
```

```python
base_dir='F:/lvcai/guangdong_round2_train_20181011/单瑕疵图片'

base_list=os.listdir(base_dir)
sec_dir=[]
for b in base_list:
    sec_dir.append(base_dir+'/'+ str(b))
```

```python
sec_dir
```

```python
def deal_points(points_list):
    #points_list = [[0, 995], [2560, 995], [2560, 1409], [0, 1409]]
    points_array=np.array(points_list)
    x=np.min(points_array[:,0])
    y=np.min(points_array[:,1])
    
    w=np.max(points_array[:,0])-x
    h=np.max(points_array[:,1])-y
    
    return x,y,w,h
```

```python
dict_class={'不导电':0,
            '擦花':1,
            '角位漏底':2,
            '桔皮':3,
            '漏底':4,
            '喷流':5, 
            '漆泡':6,
            '起坑':7,
            '杂色':8,
            '脏点':9}
```

```python
json_list=[]

for secd in sec_dir:
    sample_list=os.listdir(secd)
    for j in sample_list:
        if j[-4:]=='json':
            file_name=secd+'/'+j
            with open(file_name,encoding='UTF-8') as f:
                jsonData=json.load(f)
    
            print(jsonData['shapes'][0])  
            label=dict_class[jsonData['shapes'][0]['label']]
            x,y,w,h=deal_points(jsonData['shapes'][0]['points'])
        
            json_list.append({'file_name':file_name[41:-4]+'jpg','label':label,'x':x,'y':y,'w':w,'h':h})
```

```python
df= pd.DataFrame(json_list,columns=["file_name","label","x","y","w","h"])
df.to_csv("json_dan.csv",index=False, encoding="utf-8")
```
