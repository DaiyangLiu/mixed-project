```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
%matplotlib inline
```

```python
base_dir='F:/lvcai/guangdong_round2_train_20181011/sqaurenpy_withoutnormalize/'
npy_list=os.listdir(base_dir)
```

```python
npy_list
```

```python
length=len(npy_list)
for  i in range(0,length,2):
    print(npy_list[i])
    print(npy_list[i+1])
    print("------------")
    
    X=np.load(base_dir+npy_list[i])
    Y=np.load(base_dir+npy_list[i+1])
    
    num = X.shape[0]
    tag = int(0.8*num)
    x,y = shuffle(X,Y,random_state=2019)
    
    x_train = x[:tag,:,:,:]
    x_val = x[tag:,:,:,:]
    
    y_train = y[:tag,:,:,:]
    y_val = y[tag:,:,:,:]
    
    np.save('F:/lvcai/guangdong_round2_train_20181011/splited-squrenpy/train/X/'+str( int(i/2))+'_X_train.npy',x_train)
    np.save('F:/lvcai/guangdong_round2_train_20181011/splited-squrenpy/val/X/'+str( int(i/2))+'_X_val.npy',x_val)
    
    np.save('F:/lvcai/guangdong_round2_train_20181011/splited-squrenpy/train/Y/'+str( int(i/2))+'_Y_train.npy',y_train)
    np.save('F:/lvcai/guangdong_round2_train_20181011/splited-squrenpy/val/Y/'+str( int(i/2))+'_Y_val.npy',y_val)
```
