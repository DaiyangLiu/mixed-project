```python
import os
import cv2
import numpy as np
import pandas as pd
import PIL
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def cv_imread(filePath):
    cv_img=cv2.imdecode(np.fromfile(filePath,dtype=np.uint8),-1)
    ## imdecode读取的是rgb，如果后续需要opencv处理的话，需要转换成bgr，转换后图片颜色会变化
    ##cv_img=cv2.cvtColor(cv_img,cv2.COLOR_RGB2BGR)
    return cv_img
```

```python
def normalize_(img):
    #分别对每一层进行归一化
    new_img=np.zeros(img.shape)
    new_img[:,:,0]=(img[:,:,0]-np.mean(img[:,:,0]))/np.std(img[:,:,0])
    new_img[:,:,1]=(img[:,:,1]-np.mean(img[:,:,1]))/np.std(img[:,:,1])
    new_img[:,:,2]=(img[:,:,2]-np.mean(img[:,:,2]))/np.std(img[:,:,2])
    return new_img
```

```python
base_dir='F:/lvcai/guangdong_round2_train_20181011/多瑕疵图片'
```

```python
duo_list=os.listdir(base_dir)
```

```python
duo_list
```

```python
label_dir='F:/lvcai/guangdong_round2_train_20181011/labelsq/多瑕疵图片'
label_list=os.listdir(label_dir)
```

```python
len(label_list)
```

```python
width=480
height=480
duo_length= len(duo_list)
X_all=np.zeros( (int(duo_length/2),width,height,3) ,dtype='float64')  #(个数，长（宽width），宽(高height)，通道)
```

```python
for j in range(duo_length):
    if duo_list[j][-3:]=='jpg':
        img = cv_imread(base_dir+'/'+duo_list[j])
        img = cv2.resize(img,(width,height))
        newimg=normalize_(img)
        X_all[int(j/2),:,:,:]=newimg  #因为有json文件，所以编号要进行处理
        
np.save('F:/lvcai/guangdong_round2_train_20181011/squarenpy/duo_X.npy',X_all)
```

```python
Y_all=np.zeros( (int(duo_length/2),width,height,10) ,dtype='int8')  #(个数，长（宽width），宽(高height)，通道)

mask_length=len(label_list)
for j in range(mask_length):
    single=np.load(label_dir+'/'+label_list[j])
    Y_all[j,:,:,:]=single
    
np.save('F:/lvcai/guangdong_round2_train_20181011/squarenpy/duo_Y.npy',Y_all)
```
