#### X_TRAIN

```python
x_train_list = os.listdir(x_train_base_dir)
x_train_list
```

```python
import gc
```

```python
for i in range(11):
    print(x_train_list[i])
    if i==0:
        a=np.load(x_train_base_dir + x_train_list[0])
        b=np.load(x_train_base_dir + x_train_list[1])
        X=np.vstack((a,b))
        del a
        del b
        gc.collect()
    elif i==1:
        continue
    else:
        a=np.load(x_train_base_dir + x_train_list[i])
        X=np.vstack((X,a))
        del a
        gc.collect()
```

```python
X.shape
```

```python
np.save('F:/lvcai/guangdong_round2_train_20181011/merged-squarenpy/merged_X_train.npy',X)
```
